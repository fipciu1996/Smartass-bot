## Fundamenty, kontekst i pierwszy kanał komunikacji

**Przez najbliższe 3 dni Twoje myślenie się zmieni.** Nie dlatego, że pokażę Ci nowe narzędzie — narzędzia starzeją się szybciej, niż zdążysz je opanować. Dlatego, że zobaczysz, jak wygląda świat, w którym każdy z nas ma swojego agenta AI — "robota", który 24/7 pracuje na Twoją korzyść.

Pomyśl o tym... "Twój robot". Jak się z tym czujesz? Nie za 20 lat. Za 2–5 lat. I ten świat nie czeka, aż poczujesz gotowość. On już nadchodzi.

Jedyne pytanie brzmi: czy wejdziesz w tę zmianę świadomie — i wykorzystasz ją na swoją korzyść — czy pozwolisz, żeby Cię zaskoczyła?

Te 3 dni są po to, żeby wybrać pierwszą opcję. A sam fakt, że tu jesteś, mówi mi coś ważnego — **potrafisz rozpoznać moment, w którym warto się zatrzymać i zrozumieć, co się naprawdę dzieje.** To rzadka cecha. I bardzo Ci się teraz przyda.


### Dlaczego większość ludzi źle rozumie to, co się dzieje?

Bo myślimy liniowo, a zmiany idą **wykładniczo**.

Ludzki mózg ewoluował w świecie liniowym: 1, 2, 3, 4, 5... Wykładniczy wzrost — 1, 2, 4, 8, 16, 32, 64... — jest dla nas nieintuicyjny. Efekt? Systematycznie **przeszacowujemy** to, co będzie możliwe w najbliższych miesiącach (hype!), i **totalnie niedoszacowujemy** to, co będzie możliwe za 2–5 lat. Za 2 lata świat będzie zupełnie inny niż teraz. Za 5 lat — nie do poznania. Im szybciej to zaakceptujesz, tym lepiej dla Ciebie.

I teraz kluczowa rzecz: **nie opłaca się myśleć z perspektywy narzędzi.** Proszę, zrozum to. Narzędzie A czy B — to się szybko starzeje. Dziś OpenClaw, wczoraj LangChain, jutro coś nowego. Jeśli Twoja wiedza opiera się na konkretnym narzędziu — masz problem co kilka miesięcy.

Opłaca się myśleć z perspektywy: **jak będzie wyglądał świat i jaka będzie w nim moja rola?** No właśnie! Jaka?


### Twój robot

Na świecie już działają setki, a może tysiące fabryk produkujących roboty humanoidalne. Może być tak, że za 5 lat (lub krócej) większość z nas będzie miała swojego robota — np. za ok. 7–10 tys. zł rocznie — który będzie tworzył wartość wielokrotnie przewyższającą tę kwotę.

Ale zrozum ważną rzecz: **nie chodzi o fizycznego robota.** Chodzi o jego "mózg". O inteligencję, która pracuje na Twoją korzyść 24/7. I tę inteligencję — w formie osobistego agenta AI — możesz zacząć doświadczać **już teraz**. To nie jest obietnica za pięć lat. To jest coś, co w zarodkowej formie istnieje dzisiaj.

OpenClaw jest jednym z przykładów (nie jedynym!), który pokazuje, jak to może wyglądać. Ale to nie jest narzędzie do zainstalowania i zapomnienia. To jest **sygnał**, który mówi: zmień myślenie, bo świat właśnie zmienia reguły gry. Pomyśl o tym.


### Paradoks tego warsztatu

Będziemy mówić o OpenClaw przez 3 dni. Ale OpenClaw nie jest celem — to jest **soczewka**, przez którą zobaczysz większy obraz.

OpenClaw postawił duży nacisk na **możliwości** i bardzo mały na **kontrolę**. My w DataWorkshop Lab — pracując nad 7 Poziomami Autonomii AI — cały czas powtarzamy: im większa autonomia, tym większa musi być kontrola. Nie da się tego obejść. Nie da się.

OpenClaw zrobił dobrą robotę, bo pokazał, że pewne rzeczy są możliwe. Teraz czas, żeby zrobić to we właściwy sposób. I właśnie tego się tu nauczysz.

### Co wyczytasz z jego bloga

Polecam sprawdzić jego posty na `[steipete.me](https://steipete.me/)`. Nie po to, żeby robić psychoanalizę, tylko żeby zrozumieć, jakie przekonania wbudowały się w produkt. Wyłaniają się cztery bardzo wyraźne wzorce:

**1. Ekstremalnie niska tolerancja na tarcie.**
Peter reaguje alergicznie na wszystko, co spowalnia przejście od intencji do wykonania: zbyt grube frameworki, zbyt ceremonialne procesy, zbyt "sprytne" warstwy abstrakcji. To wyjaśnia, dlaczego OpenClaw daje agentowi bezpośredni dostęp do terminala zamiast budować pięć warstw pośredniczących.

**2. Abstrakcje muszą zarabiać na swoje istnienie.**
W starszym, ostrym wpisie o SwiftUI (*"The State of SwiftUI"*) krytykuje nie nowość technologii, ale fakt, że zaslania mechanikę i utrudnia przewidywanie. Dokładnie ten sam odruch wraca przy AI: sceptycyzm wobec zbyt rozbudowanego MCP (Model Context Protocol), preferencja dla CLI i prostych narzędzi systemowych.

**3. Sprawczość ponad elegancję.**
Peter mierzy technologie jednym pytaniem: *"Czy to mi realnie pozwala szybciej zbudować, sprawdzić i dostarczyć to, co mam w głowie?"* Nie: "Czy to jest eleganckie akademicko?" Nie: "Czy to jest modne?" To dlatego jest jednocześnie wczesnym adopterem *i* bezlitosnym krytykiem niedojrzałych narzędzi.

**4. Sygnał ponad szum.**
W starszych wpisach (*"How to Use Slack and Not Go Crazy"*, *"Gardening Your Twitter"*) widać tego samego człowieka co dziś: kogoś, kto walczy o czystość kanału informacyjnego. Dawniej kuratorował Slacka i Twittera. Dziś kuratoruje kontekst dla agentów AI. Ta sama walka, inny front.

## Co to jest OpenClaw?


### Nie chatbot - agent

Kiedy słyszysz "AI", pewnie myślisz o ChatGPT. Wpisujesz pytanie → dostajesz odpowiedź → koniec.

OpenClaw działa inaczej:

| Tradycyjne AI (ChatGPT) | OpenClaw |
|---|---|
| Pytasz → odpowiada | Konfigurujesz raz → działa autonomicznie |
| Rozmowa kończy się po każdej wymianie | Działa 24/7, pamięta kontekst |
| Reaktywne — czeka na Ciebie | Proaktywne — samo inicjuje działania |
| Nie ma dostępu do systemu | Ma dostęp do maili, kalendarza, plików, terminala, przeglądarki |
| Statyczne możliwości | **Samo się rozbudowuje** — instaluje sobie nowe narzędzia |

Przeczytaj ten ostatni punkt jeszcze raz. OpenClaw potrafi **modyfikować własną konfigurację**. Poprosisz go: *"Chcę, żebyś sprawdzał moje maile co godzinę i wysyłał mi podsumowanie na Telegram"* — a on sam czyta swoją dokumentację, doinstalowuje moduły, edytuje pliki konfiguracyjne, restartuje się i zaczyna działać.

### Saga nazewnicza

### Koncepcja "Twojego agenta"

To jest kluczowe słowo: **Twój**. Tak, tak - Twój! Nie chatbot na stronie — *Twój* agent, który 24/7 pracuje na Twoją korzyść. Kombinuje, usprawnia, pilnuje, przypomina.

Ludzie, którzy tego doświadczyli — kiedy agent naprawdę podpowiada coś, co masz zrobić, albo wyłapuje rzecz, która by Ci uciekła — opisują to jako przeskok, którego nie da się przekazać słowami. Trzeba go **doświadczyć**.

Właśnie dlatego robimy ten DWthon. Żeby dać Ci możliwość doświadczyć i zrozumieć, że świat zmienia się szybciej niż myślisz. Czas zdobywać nowe kompentecji. A na koniec dzisiejszego dnia postawisz pierwszy element: kanał komunikacji na Twoim telefonie.


## Software 3.0 — Trzy światy programowania

Żeby zrozumieć, *dlaczego* OpenClaw jest jednocześnie fascynujący i niebezpieczny — i czemu większość ludzi źle interpretuje tę rewolucję — musisz najpierw zrozumieć jedną fundamentalną koncepcję. Stoję na przecięciu trzech światów — i to w DataWorkshop nazywamy **Software 3.0**.

### Software 1.0: Klasyczne IT

To świat, który znasz:
- Kod pisany ręcznie przez programistę
- Reguły `if/else`, algorytmy, bazy danych
- **Deterministyczny**: te same dane wejściowe → zawsze ten sam wynik
- Przewidywalny, nudny, ale *niezawodny*

```python
if temperatura > 30:
    print("Włącz klimatyzację")
else:
    print("Nic nie rób")
```

Programista decyduje o każdym kroku. Komputer wykonuje rozkazy.

### Software 2.0: Klasyczne Machine Learning

Tu model **uczy się z danych** zamiast być programowany ręcznie:
- Sieci neuronowe, drzewa decyzyjne, regresja
- Model trenowany na przykładach → "uczy się" wzorców
- **Niedeterministyczny**: model popełnia błędy. Zawsze

Tu nauczyłem się pokory. W klasycznym ML jedną z pierwszych lekcji jest: **model ZAWSZE popełnia błąd**. Pytanie nie brzmi "czy się pomyli?", tylko "jak często i jak bardzo?".

### Software 3.0: Programowanie językiem naturalnym

To nowy świat:
- Model **rozumie** język naturalny
- Możesz wydać polecenie słowami: *"Przeanalizuj to CV i wyciągnij doświadczenie w Pythonie"*
- **Potężny**, bo eliminuje konieczność pisania reguł dla każdego przypadku
- **Niedeterministyczny do potęgi**: nigdy nie masz 100% pewności, co model zwróci

I tu dochodzimy do sedna — dlaczego większość ludzi źle rozumie tę rewolucję.

### Pułapka "szybszych koni"

Henry Ford ponoć powiedział: *"Gdybym zapytał ludzi, czego chcą, powiedzieliby: szybszych koni."*

Obserwuję to samo w 2026 roku. Większość ludzi, którzy "używają AI", robi dokładnie jedno: **używa Claude Code albo Codex, żeby szybciej generować Software 1.0**.

Piszą prompty: *"Napisz mi funkcję w Pythonie, która parsuje CSV"*. AI generuje kod. Szybciej niż ręcznie. Super.

Ale to jest **szybszy koń**. Nie samochód.

Umiejętność jutra nie polega na tym, żeby szybciej tworzyć stare rzeczy. Polega na tworzeniu **nowych rzeczy w nowy sposób**. Projektowaniu systemów, w których AI ma swoje miejsce, narzędzia, uprawnienia i ramy kontroli. To jest zasadniczo inny paradygmat.

Kiedy Ford budował samochody, ludzie dziwili się: *"Jak to — jeździć czymś, co nie je owsa?"*. Wydawało się to nienaturalne, dziwne, niebezpieczne. A potem stało się oczywiste.

Z AI jest tak samo. Dziś ludzie myślą: *"AI pisze za mnie kod"* (szybszy koń). Jutro będą myśleć: *"AI jest częścią systemu, który rozwiązuje problemy, jakich nie mogłem rozwiązać sam"* (samochód).

## Harness Engineering — nowe rzemiosło Software 3.0

Każdy paradygmat programowania ma swoje rzemiosło:

- **Software 1.0** miał algorytmy, struktury danych, clean code, wzorce projektowe
- **Software 2.0** miał feature engineering, dobór modeli, ewaluację, zarządzanie danymi treningowymi
- **Software 3.0** ma… co?

Przez chwilę branża myślała, że odpowiedź brzmi **Prompt Engineering**. Napisz lepszy prompt → dostaniesz lepszy wynik. To był dobry start, ale za mało.

Potem pojawiło się **Context Engineering** — nie chodzi tylko o to, co powiesz modelowi, ale jaki kontekst mu dostarczysz (dane, dokumenty, historię). To już było bliżej.

Ale teraz widzimy coś szerszego. Zarówno OpenAI, jak i Anthropic zaczynają mówić o czymś, co się krystalizuje pod nazwą **Harness Engineering**.

### Co to jest?

Wyobraź sobie, że zatrudniasz nowego pracownika. Nie wystarczy powiedzieć mu "pracuj dobrze" (prompt). Nie wystarczy nawet dać mu dostęp do firmowej bazy wiedzy (context). Musisz zaprojektować:

- **Jakie narzędzia** dostanie? (laptop, dostęp do CRM, klucze do biura)
- **Jakie ma uprawnienia?** (może przeglądać dane klientów, ale nie kasować)
- **Skąd bierze informacje?** (konkretne foldery, nie "cały internet")
- **Jak mierzysz jego pracę?** (KPI, review, feedback)
- **Co się dzieje, gdy się pomyli?** (kto sprawdza, jaki jest blast radius)
- **Jak wygląda jego "biurko"?** (czyste, uporządkowane, ze wszystkim pod ręką)

**Harness Engineering to projektowanie całego tego środowiska — ale dla agenta AI.**

Nie "co powiedzieć modelowi", tylko: **w jakim systemie model ma działać, z jakimi granicami, z jaką informacją zwrotną i jak ograniczyć szkody, kiedy się pomyli**.

OpenAI pisze o tym wprost: *"Humans steer. Agents execute."* Ludzie kierują. Agenty wykonują. Ale żeby agent wykonywał dobrze, musisz zaprojektować cały układ sterowania.

### 7 warstw harnessu

W praktyce dobrze zaprojektowany harness ma co najmniej siedem warstw:

1. **Intent** - cele, ograniczenia, instrukcje, kryteria "zrobione"
2. **Context** - jakie pliki, dokumenty, logi, dane agent dostaje
3. **Tools** — jakie narzędzia ma do dyspozycji i jak są zaprojektowane
4. **Permissions** — co wolno zrobić automatycznie, co wymaga zgody, co jest zabronione
5. **Eval** — jak mierzysz jakość, regresję, bezpieczeństwo
6. **Runtime** — sesje, stan, retries, izolacja, cleanup, observability
7. **Governance** — wersjonowanie, docs jako source of truth, audyt

### Dlaczego to jest Software 3.0, a nie "lepszy prompt"?

Bo w Software 1.0 Twoja praca to **pisanie kodu**. W Software 3.0 Twoja praca to **projektowanie środowiska, w którym AI operuje**. To jest fundamentalna zmiana roli.

Ktoś, kto używa Claude Code do generowania funkcji w Pythonie, dalej robi Software 1.0 — tylko szybciej. **Szybszy koń.**

Ktoś, kto projektuje harness: definiuje narzędzia, uprawnienia, ewaluację i pętle zwrotne dla agenta — robi Software 3.0. **Samochód.**

Obie umiejętności mają wartość. Ale tylko jedna jest przyszłością.

I teraz: OpenClaw jest dobrym przykładem tego, co się dzieje, gdy budujesz potężnego agenta (mocny silnik), ale harness (cały układ sterowania) jest jeszcze niedojrzały. Rezultat? Impresywne demo, ale też skasowane skrzynki pocztowe i 42 tysiące instancji wystawionych na publiczny internet bez autentykacji. O tym więcej jutro.

Na razie zapamiętaj tę zasadę:

> **Agent AI bez harnessu to silnik bez hamulców.**

## 7 Poziomów Autonomii AI

W DataWorkshop używamy autorskiej mapy, która porządkuje chaos w głowie. Pozwala odpowiedzieć na pytanie, które wielu ludzi (i firm) zadaje źle: *"Na jakim etapie jesteśmy i co mamy robić dalej?"*

Wyobraź to sobie jako suwak:
- Z jednej strony masz **Human-in-the-Loop (100%)** — człowiek sprawdza każdy krok
- Z drugiej masz **Pełną Autonomię (0%)** — maszyna o wszystkim 100% decyduje sama (na razie *science fiction*)


<img src="../images/7_level_ai.jpg" width="1000px" />


### 🟢 FAZA 1: FUNDAMENTY

#### Poziom 1: Konwersacyjne AI (Chat)
- To jest **99% rynku**. "Napisz maila", "Opowiedz żart" — ChatGPT, Claude, Gemini
- Model "gada", ale nie "działa". Zerowa powtarzalność
- Nie da się tego wdrożyć na produkcję. To zabawka, nie system

Istnieje mnóstwo materiałów o tym, jak pisać prompty *ad hoc*. W DataWorkshop ten poziom pomijamy — nie dlatego, że jest zły (a może i dlatego 😉), ale dlatego, że nie da się na nim zbudować niczego, za co ktoś może wziąć odpowiedzialność.

#### Poziom 2: Programowalne AI (Structured Output)
- **Fundament wszystkiego, co dalej**
- Model przestaje lać wodę. Zwraca **strukturę** (JSON, obiekty)
- Dzięki temu możesz podpiąć go pod kod w Pythonie. AI staje się elementem systemu IT
- Od tego poziomu zaczyna się realna wartość biznesowa — i wiele firm tego jeszcze nie dostrzegło

W języku Harness Engineering: na Poziomie 2 zaczynasz projektować **kontrakt** między Tobą a modelem. Definiujesz, co model *musi* zwrócić. To pierwszy element harnessu.

### 🟡 FAZA 2: POŁĄCZONE AI

#### Poziom 3: Zintegrowane AI (RAG)
- Model dostaje dostęp do Twojej bazy wiedzy. Szuka i cytuje informacje
- Żeby RAG działał dobrze, najpierw musisz opanować Poziom 2 — bo model musi umieć czytać i zwracać strukturę
- *(Uczymy głębiej w: [Praktyczny LLM](https://dataworkshop.eu/pl/praktyczny-llm?utm_source=dwthon-agentic))*

#### Poziom 4: Kontekstowe AI (Pamięć + Narzędzia) 🧠
- Model zyskuje **Pamięć** (pamięta kontekst między sesjami) i **Narzędzia** (może wywołać API, wysłać maila, otworzyć przeglądarkę)
- Tu zaczyna się prawdziwa "agencja" — tu agent zaczyna *działać*

#### Poziom 5: Strategiczne AI (adaptywne planowanie)
- System nie tylko ma narzędzia, ale potrafi **planować** i adaptować plan na bieżąco
- *"Muszę sprawdzić pogodę, żeby wiedzieć, czy wysłać ofertę na parasole"*
- *(Poziomy 4 i 5 to domena: [Agentic AI](https://dataworkshop.eu/agentic-ai?utm_source=dwthon-agentic) — start: 30 marca)*

#### Poziomy 6-7: Meta-autonomia i pełna autonomia (AGI)
Na razie teoretyczne punkty odniesienia. Do 2030 nie widzę trajektorii dla AGI (czyli poziom 7my).

### 🦞 Gdzie jest OpenClaw na tej mapie?

Gdybyś zapytał marketing OpenClaw, odpowiedziałby: *"Poziom 5! Agent planuje, używa narzędzi, sam się rozbudowuje!"*

Gdybyś zapytał mnie, odpowiedziałbym precyzyjniej:

> **OpenClaw celuje w Poziom 4–5, ale ma poważne dziury na Poziomie 2.**

Co to znaczy? OpenClaw ma pamięć (Poziom 4). Ma narzędzia (Poziom 4). Ma elementy planowania (Poziom 5). Potrafi zainstalować sobie nowe umiejętności (meta-poziom). Ale jego fundamenty — kontrola struktury wyjścia, walidacja odpowiedzi, zarządzanie typami danych — są **niedojrzałe**. LLM halucynuje? W ChatGPT to jest zabawne. W OpenClaw, który edytuje plik konfiguracyjny JSON, zapomniany przecinek **wyłącza cały system**.

To jest dokładnie ten wzorzec, przed którym ostrzegamy od początku: **nie przeskakujesz poziomów.**

Większość projektów AI kończy się fiaskiem, bo biznes oczekuje poziomu 5 (bo zobaczył demo OpenClaw), a zespół nie umie wymusić na modelu zwrócenia poprawnego JSON-a (Poziom 2).

W języku Harness Engineering: OpenClaw ma mocny silnik (capabilities), ale harness (układ sterowania) jest jeszcze w budowie.

### ⚠️ Największy mit: "Agent AI"

Uważaj na pułapkę językową. Słowo "Agent AI" **nie** oznacza poziomu zaawansowania.

Możesz mieć:
- Głupiego agenta na **Poziomie 1** (chatbot z instrukcją "Jesteś piratem")
- Systemowego agenta na **Poziomie 2** (wyciąga dane do struktury JSON)
- Strategicznego agenta na **Poziomie 5** (planuje i wykonuje złożone zadania)

Kiedy ktoś mówi "wdrożyłem Agenta AI", nie klaskaj od razu. **Zapytaj: na którym poziomie?**

## Paradoks Kontroli: żelazna zasada inżynierii AI

To jest najważniejsza zasada, którą kierujemy się w DataWorkshop:

> **Im wyższy poziom autonomii, tym wyższa musi być kontrola.**

Często słyszę: *"Chcę dać AI autonomię, żeby nie musieć go pilnować."* To jest **błąd logiczny.**

Gdy dajesz agentowi dostęp do narzędzi (API, plików, maili, terminala, przeglądarki), ryzyko rośnie wykładniczo. Dlatego musisz **zwiększyć** kontrolę. Ale są dwa rodzaje kontroli:

1. **Kontrola manualna (human-in-the-loop):** Człowiek zatwierdza każdy krok. Działa na początku, ale **nie skaluje się**.

2. **Kontrola systemowa (harness):** Przemyślana architektura — walidacja typów, twarde ograniczenia uprawnień, izolacja środowisk. System, który **uniemożliwia** pewne działania.

Zapamiętaj: **im wyższy poziom autonomii, tym mniej człowieka w procesie, ale WIĘCEJ człowieka w architekturze.** Cały wysiłek wkładasz *przed* uruchomieniem, projektując harness.

OpenClaw jest idealną ilustracją tego, co się dzieje, gdy tę zasadę nagiąsz. Autorzy świadomie postawili na **capability first** — najpierw moc, potem bezpieczeństwo. Dzięki temu odkryli, że rynek naprawdę chce autonomicznych agentów. Ale konsekwencje? O tym porozmawiamy szczegółowo jutro — bo ten temat zasługuje na osobny dzień.

Na razie jedna liczba, żebyś miał skalę problemu:

### 💰 Ekonomia tokenów — ile to naprawdę kosztuje?

OpenClaw potrafi być **bardzo tokenożerny**. Nie zawsze — bo dużo zależy od konfiguracji. Ale domyślnie każda interakcja wysyła do LLM spory kontekst: system prompt, definicje dostępnych narzędzi, instrukcje agenta, historię rozmowy, listę umiejętności (skills). Do tego dochodzą mechanizmy, które działają w tle: **heartbeat** (cykliczne "budzenie" agenta), **subagenty** (wyspecjalizowani pomocnicy wywoływani automatycznie) czy scheduled tasks. Każdy z tych elementów to dodatkowe tokeny — nawet jeśli akurat nie ma nic do zrobienia.

To wszystko da się konfigurować: możesz wyłączyć heartbeat, ograniczyć subagentów, skrócić kontekst. Ale domyślna konfiguracja — ta, z którą startuje większość ludzi — jest hojną w zużyciu.

Weźmy jako przykład cennik **Claude Opus 4.6** (model, który OpenClaw domyślnie proponuje):

- **Input:** `$5` / 1M tokenów
- **Output:** `$25` / 1M tokenów

| Przykładowa akcja | Tokeny (rząd wielkości) | Szacowany koszt |
|---|---|---|
| Restart sesji ("cześć") | ~10 000 in + ~500 out | ~\$0,06 |
| Ustaw przypomnienie | ~45 000 in + ~2 000 out | ~\$0,28 |
| Dodaj 2 strony do projektu | ~263 000 in + ~5 000 out | ~\$1,44 |
| Złożone zadanie kodowania (25 min) | ~1 000 000+ in + ~50 000 out | **\$6–7+** |

> ☝️ To są **przykładowe** wartości. Twoje realne zużycie będzie zależeć od modelu, konfiguracji, liczby narzędzi i długości historii rozmów. Ale rząd wielkości się nie zmieni drastycznie.

Teraz prosta matematyka, której nikt w hype'owych postach nie robi:

- Przy intensywnym użytkowaniu (np. 50 interakcji dziennie z aktywnym heartbeatem): **kilkaset dolarów dziennie** to realistyczny scenariusz
- Miesięcznie to mogą być **tysiące, a nawet dziesiątki tysięcy dolarów**
- W takim scenariuszu koszt agenta **może przekroczyć pensję pracownika**, którego rzekomo zastępuje

Czy to znaczy, że OpenClaw jest bezużyteczny? Nie. Znaczy, że trzeba **świadomie zarządzać konfiguracją**, wybrać odpowiedni model do zadania (nie zawsze potrzebujesz Opusa — tańsze modele radzą sobie z prostszymi zadaniami) i rozumieć, za co płacisz. To jest właśnie Harness Engineering w praktyce: nie tylko "co agent robi", ale "ile to kosztuje i jak to zoptymalizować".

Hasło "zwolniłem 15 pracowników" brzmi sexy na LinkedIn. Ale wystarczy kalkulator, żeby zobaczyć, że — przy dzisiejszych cenach i domyślnych konfiguracjach — jest to w najlepszym przypadku uproszczenie. To się zmieni: modele tanieją, architektura agentów dojrzewa, optymalizacje kontekstu wchodzą. Ale **dzisiaj** — warto znać te liczby, zanim podejmiesz decyzję.

Swoją drogą — sprawdzaliśmy to w naszym Labie. Postawiliśmy OpenClaw, robiliśmy dość proste testy, żadne skomplikowane scenariusze. W kilka dni nagrzał ponad $200 😱. Kiedy przeanalizowaliśmy logi, okazało się, że ponad 95% zużycia to nie była produktywna praca — to był heartbeat, subagenty gadające ze sobą, utrzymywanie kontekstu i cykliczne "budzenia". Agent głównie rozmawiał sam ze sobą. Za nasze pieniądze 🙂.

## Co jest w OpenClaw naprawdę wartościowe?

Po tej dawce pragmatyzmu mogłoby się wydawać, że OpenClaw to zło wcielone. **Nie.** OpenClaw jest ważny — ale nie z powodów, o których mówi internet.

**1. Pokazał prawdziwy popyt.** Ludzie nie chcą kolejnej zakładki z AI. Chcą asystenta, który siedzi tam, gdzie oni żyją: w komunikatorze, w telefonie.

**2. Pozwolił doświadczyć "twojego agenta".** Koncepcja agenta, który 24/7 pracuje na Twoją korzyść, przestała być abstrakcją. Ludzie to przeżyli. A kiedy przeżyjesz — to zmienia myślenie w sposób, którego nie da się cofnąć.

**3. Potwierdził Paradoks Kontroli.** Ma rację produktowo (kierunek jest właściwy) i jednocześnie nie rozwiązuje centralnego problemu bezpieczeństwa. Dokładnie to, o czym mówimy od lat.

**4. Zmienił umysły.** Przed OpenClaw wielu myślało o AI jak o "lepszym Google". Po OpenClaw ludzie rozumieją, że AI może mieć **sprawczość**. Za to mu cześć 🦞.

### Pamiętaj o DynaTAC

OpenClaw jest "cegłą". Ciężki, drogi, z problemami. Ale DynaTAC też był "cegłą" — a potem zmienił świat. Kierunek jest właściwy. Pytanie brzmi: jak to zrobić **bezpiecznie i optymalnie**? Tym zajmujemy się w DataWorkshop.

### Krok 3: Sprawdź, czy bot żyje

````python
response = requests.get(f"{BASE_URL}/getMe")
bot_info = response.json()

if bot_info.get("ok"):
    bot = bot_info["result"]
    print(f"✅ Bot działa!")
    print(f"   Imię:     {bot['first_name']}")
    print(f"   Username: @{bot['username']}")
else:
    print("❌ Coś poszło nie tak. Sprawdź token.")
    print(bot_info)
````

### Krok 4: Wyślij wiadomość z telefonu

Teraz:
1. Otwórz Telegram na telefonie
2. Wyszukaj swojego bota po username (ten z `@`)
3. Wyślij mu wiadomość, np.: **`hello`**

Wróć tu po chwili — w następnej komórce odbierzemy tę wiadomość.

### Krok 5: Odbierz wiadomość w kodzie

````python
# Pobieramy wiadomości, które ktoś (Ty!) wysłał do bota
response = requests.get(f"{BASE_URL}/getUpdates")
updates = response.json()

if updates.get("ok") and updates["result"]:
    last_update = updates["result"][-1]
    message = last_update.get("message", {})
    
    chat_id = message.get("chat", {}).get("id")
    text = message.get("text", "")
    sender = message.get("from", {}).get("first_name", "???")
    
    print(f"📨 Wiadomość od: {sender}")
    print(f"   Tekst:       {text}")
    print(f"   Chat ID:     {chat_id}")
    print(f"\n💡 Zapamiętaj chat_id — za chwilę wyślemy odpowiedź z powrotem na telefon!")
else:
    print("📭 Brak wiadomości. Upewnij się, że wysłałeś coś do bota z telefonu.")
    print("   Spróbuj ponownie za chwilę.")
````

### Krok 6: Wyślij odpowiedź z kodu na telefon

Teraz zamykamy pętlę. Wiadomość przyszła z telefonu → kod ją odebrał → teraz **kod odpowiada na telefon**.

````python
# Wysyłamy wiadomość z kodu na Twój telefon
response = requests.post(
    f"{BASE_URL}/sendMessage",
    json={
        "chat_id": chat_id,
        "text": "🦞 Cześć z DWthon! Twój kanał komunikacji działa. Jutro popłynie tędy inteligencja.. Chyba, trzymajmy kciuki :)"
    }
)

if response.json().get("ok"):
    print("✅ Wiadomość wysłana! Sprawdź telefon.")
else:
    print("❌ Nie udało się wysłać. Sprawdź chat_id.")
````

### Krok 7: Echo bot — zamykamy Hello World 🎉

Na koniec — klasyczny echo bot. Wyślesz `hello`, dostaniesz `olleh`. To udowadnia, że **cała rura działa**: telefon → internet → Twój kod → internet → telefon.

Bot podziała 60 sekund, a potem się zatrzyma (żebyśmy nie blokowali notebooka).


````python
def run_echo_bot(duration_seconds=60):
    """Echo bot — odwraca tekst. Działa przez podany czas."""
    offset = None
    end_time = time.time() + duration_seconds
    
    print(f"🦞 Echo bot działa przez {duration_seconds} sekund.")
    print("   Wyślij wiadomość do bota z telefonu!\n")
    
    while time.time() < end_time:
        params = {"timeout": 5}
        if offset is not None:
            params["offset"] = offset
            
        try:
            response = requests.get(f"{BASE_URL}/getUpdates", params=params, timeout=10)
            
            for update in response.json().get("result", []):
                offset = update["update_id"] + 1
                message = update.get("message", {})
                text = message.get("text", "")
                msg_chat_id = message.get("chat", {}).get("id")
                
                if text and msg_chat_id:
                    reversed_text = text[::-1]
                    requests.post(
                        f"{BASE_URL}/sendMessage",
                        json={"chat_id": msg_chat_id, "text": reversed_text}
                    )
                    print(f"   📨 Otrzymano: {text}  →  📤 Wysłano: {reversed_text}")
                    
        except requests.RequestException as e:
            print(f"   ⚠️ Błąd: {e}")
            time.sleep(2)
    
    print("\n⏱️ Bot zatrzymany.")

# Uruchom echo bota na 60 sekund
run_echo_bot(60)
````

### Co udało się zrobić?

Zbudowałeś **rury**. Kanał komunikacji między Twoim telefonem a kodem na serwerze. Na razie płynie tędy `hello` → `olleh`. To jest jak podłączenie wody do domu — sama woda jeszcze nie jest ciepła, ale instalacja działa.

Jutro do tego kanału podłączymy **inteligencję**: LLM, narzędzia, przeglądarkę. I wtedy zamiast `olleh` dostaniesz prawdziwe odpowiedzi. Agent zacznie *myśleć*.

Ale fundament jest gotowy. Masz go na telefonie. Działa.
